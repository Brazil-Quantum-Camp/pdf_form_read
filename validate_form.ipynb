# 📄 Verificação do Arquivo PDF – Lista de Exercícios

Este notebook serve para verificar se o **arquivo PDF da lista de exercícios** pode ser **lido corretamente pelo sistema de avaliação**.

Dependendo de **como o PDF é salvo**, o sistema pode não conseguir identificar corretamente as respostas. Por isso, é **muito importante fazer esta verificação antes de enviar a lista**.


## ▶️ Como usar este notebook

1. Execute a **primeira célula** do notebook.
2. Quando solicitado, **carregue o arquivo PDF da lista de exercícios**.
3. Em seguida, execute a **próxima célula** para realizar a verificação.


## 📊 O que você deve ver ao final

Se o PDF estiver correto, será exibida uma **tabela** com:

* Seus **dados pessoais**
* As **respostas preenchidas** em cada questão

Confira com atenção se todas as informações estão corretas.

⚠️ **Se a tabela não aparecer**, isso significa que **o sistema não conseguiu ler o PDF**.
Nesse caso, preencha novamente os campos da lista e **salve o arquivo corretamente**.

> ❗ **Não utilize a opção "Imprimir para PDF"**.
> Para salvar, use **Ctrl + S** ou a opção **Salvar**.


## ℹ️ Informações importantes sobre as respostas

* Este notebook **não corrige a lista** e **não atribui nota**.
  Ele apenas verifica se o PDF foi **salvo de forma que o sistema consiga ler as respostas**.
* Em **questões de múltipla escolha**, a tabela deve mostrar a letra escolhida
  (`A`, `B`, `C` ou `D`).
* Em **questões numéricas**, deve aparecer um **número**.
* Em **questões de código**, a tabela mostrará duas informações:
  1. `True` ou `False`, indicando se o código está **sintaticamente correto**
     (apenas a sintaxe é verificada; o código **não é executado**)
  2. Uma **string com o código digitado**
  > Observação: o código aparece em **uma única linha**, usando `\n` para indicar quebras de linha.
* Se alguma resposta aparecer como `None`, isso significa que:
  * Em questões de **múltipla escolha**, nenhuma opção foi selecionada
  * Em questões **numéricas**, a resposta não pôde ser convertida para um número
